## Data Modeling

### 1. Logistic Regression

```python
test_scoresl = ['test_scores']
train_scoresl = ['train_scores']
for i in range(1,10):
    cls0 = df[df['Class']==0]
    cls0 = cls0.sample(n=492, random_state=i)
    df0 = pd.concat([cls0,cls1])
    X = df0.drop(['Class'],axis=1)
    scaler = preprocessing.StandardScaler().fit(X)
    X_scaled=scaler.transform(X) 
    y= df0['Class'].ravel()
    X_train,X_test,y_train,y_test= train_test_split(X_scaled, y, test_size=0.25, random_state=1)
    logreg = LogisticRegression()
    logreg.fit(X_train, y_train)
    train_scoresl.append(round(logreg.score(X_train, y_train),4))
    test_scoresl.append(round(logreg.score(X_test, y_test),4))
for i in zip(train_scoresl,test_scoresl):
    print(i)
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,10),train_scoresl[1:],marker='*',label='Train Score')
plt.ylim(0.5,1)
plt.ylabel('Scores')
plt.title("Logistic Regression Learning Curve")
p = sns.lineplot(range(1,10),test_scoresl[1:],marker='o',label='Test Score')
```

```python
cls0 = df[df['Class']==0]
cls0 = cls0.sample(n=492, random_state=4)
df0 = pd.concat([cls0,cls1])
X = df0.drop(['Class'],axis=1)
scaler = preprocessing.StandardScaler().fit(X)
X_scaled=scaler.transform(X) 
y= df0['Class'].ravel()
X_train,X_test,y_train,y_test= train_test_split(X_scaled, y, test_size=0.25, random_state=1)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y_train,X_train)
result=logit_model.fit()
print(result.summary2())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print('Accuracy of svm classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

### 2. SVM

```python
from sklearn import svm
```

```python
regr = svm.SVC(kernel='rbf')
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
```

```python
print('Accuracy of svm classifier on test set: {:.2f}'.format(regr.score(X_test, y_test)))
```

```python
test_scoresvm = ['test_scores']
train_scoresvm = ['train_scores']
for i in range(1,10):
    cls0 = df[df['Class']==0]
    cls0 = cls0.sample(n=492, random_state=i)
    df0 = pd.concat([cls0,cls1])
    X = df0.drop(['Class'],axis=1)
    scaler = preprocessing.StandardScaler().fit(X)
    X_scaled=scaler.transform(X) 
    y= df0['Class'].ravel()
    X_train,X_test,y_train,y_test= train_test_split(X_scaled, y, test_size=0.25, random_state=1)
    svm_ = svm.SVC(kernel='rbf')
    svm_.fit(X_train, y_train)
    train_scoresvm.append(round(svm_.score(X_train, y_train),4))
    test_scoresvm.append(round(svm_.score(X_test, y_test),4))
for i in zip(train_scoresvm,test_scoresvm):
    print(i)
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,10),train_scoresvm[1:],marker='*',label='Train Score')
plt.ylim(0.5,1)
plt.ylabel('Scroes')
plt.title("SVM Learning Curve")
p = sns.lineplot(range(1,10),test_scoresvm[1:],marker='o',label='Test Score')
```

```python
svm_ = svm.SVR()
svm_.fit(X_train, y_train)
train_scoresvm.append(round(svm_.score(X_train, y_train),4))
test_scoresvm.append(round(svm_.score(X_test, y_test),4))
```

## 3. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
test_scoresk = ['test_scores']
train_scoresk = ['train_scores']
for i in range(1,10):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train, y_train)
    train_scoresk.append(round(knn.score(X_train,y_train),4))
    test_scoresk.append(round(knn.score(X_test,y_test),4))
```

```python
for i in zip(test_scoresk,train_scoresk):
    print(i)
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,10),train_scoresk,marker='*',label='Train Score')
p = sns.lineplot(range(1,10),test_scoresk,marker='o',label='Test Score')
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors':np.arange(1,50)}
knn = KNeighborsClassifier()
knn_cv= GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(X,y)
```

```python
print("Best Score:" + str(knn_cv.best_score_))
print("Best Parameters: " + str(knn_cv.best_params_))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
